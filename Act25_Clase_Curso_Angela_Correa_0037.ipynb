```python
class Curso:
    # El método __init__ inicializa los atributos de la clase
    def __init__(self):
        self.nombre = ""
        self.duracion = ""
        self.costo = 0.0

    # Función para pedir datos al usuario (Interactivo)
    def leer_datos(self, numero_objeto):
        print(f"\n--- Ingresando datos para el Curso {numero_objeto} ---")
        self.nombre = input("Nombre del curso: ")
        self.duracion = input("Duración (ej. 20 horas): ")
        self.costo = input("Costo del curso: ")
        print(f"✅ ¡Datos del curso '{self.nombre}' guardados con éxito!")

    # Función para mostrar los datos en pantalla
    def mostrar_datos(self):
        print(f"\n> INFORMACIÓN DEL CURSO: {self.nombre}")
        print(f"  - Duración: {self.duracion}")
        print(f"  - Inversión: ${self.costo}")

# --- Bloque Principal ---

# 1. Creamos los 2 objetos (instancias de la clase Curso)
curso1 = Curso()
curso2 = Curso()

# 2. Llamamos a la función leer_datos para cada objeto
curso1.leer_datos(1)
curso2.leer_datos(2)

# 3. Llamamos a la función mostrar_datos para ver el resultado
print("\n" + "="*30)
print("  RESUMEN DE CURSOS CREADOS")
print("="*30)

curso1.mostrar_datos()
curso2.mostrar_datos()
print("Programa realizado por Angela Michelle Correa Lopez NL = 13 ")
```
